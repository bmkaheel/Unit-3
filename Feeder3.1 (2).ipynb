### Making Use of our Data

Although all the data in our raw dataset is useful, in our case, it is not. In a previous exercise, we used our data and came up with three different analysis areas and made our research questions. Looking at the whole dataset can be quite overwhelming so we will try to narrow it down step-by-step in an effort to make a subset of data, making it easier to answer those questions. But first, let's take a look at our three analysis areas. 

#### Analysis Area #1

In the **first** analysis area, we decided that there are areas that would benefit from the addition and expansion of healthcare and healthcare coverage. A reason for low health can be attributed to the cost of treatment. Healthcare can be very heavy on a households expenses and some may choose to deal with their inadequate health rather than receive treatment.

#### Analysis Area #2

In the **second** analysis area, we decided that there are areas that would benefit from an increased tobacco tax. More specifically, areas that would benefit health-wise from an increased tobacco tax. This would be an effort to restrict the use of tobacco and smoking.

#### Analysis Area #3

In the **third** analysis area, we decided that there are areas that would benefit from increased funding. Some people live in areas that are below the poverty line. This can lead to negative health outcomes in different ways such as unhealthy foods or inadequate housing.


### Analysis Area Subset 1

This **first** subset can be used to provide more detail on the first analysis area regarding hospitals costs payments but specially to the State of North Carolina. The columns selected limit the data to **ONLY** North Carolina. We are able to then see the which counties have which rates of uninsured individuals and how much hospital costs are. By doing this, we are able to compare the different counties with the State of North Carolina and correlating their health data to the impact of cost as a reason people couldn't receive treatment. 

````python
raw_data.loc[3243:3442,['State', 'County', 'FIPS',  'Uninsured', 'Health care costs', 'Could not see doctor due to cost']]
````

````python
Hospital_Costs = raw_data.loc[3243:3442,['State', 'County', 'FIPS',  'Uninsured', 'Health care costs', 'Could not see doctor due to cost']].copy()
````

````python
Hospital_Costs.to_csv("Hospital_Costs.csv", index=False)
````

### Analysis Area Subset 2

The **second** subset above can be used to provide more detail on the second analysis area regarding a tax impose tobacco and alcohol consumption. Smoking and alcohol-use has long be correlated to negative health and this is a contributing factor to the negative health rates of counties, some more than others. We can see from the data set below the rates at which adults engage in either smoking or excessive drinking and this habits, if stopped or even limited, can increase the health rates of these counties immensely and this is can be done by imposing an additional tax on those products in an attempt to limit the purchasing of them. Another idea that may help people get away from the use of tobacco and alcohol could be by limiting their free time and giving them something else to do. This could be done by proposing a tax break for exercise and fitness related products and activities bring individuals to healtier habits rather than detremental ones. 

````python
raw_data.loc[3243:3442,['State', 'County', 'FIPS', 'Adult smoking', 'Excessive drinking', 'Alcohol-impaired driving deaths']]
````

````python
Alcohol_tobacco = raw_data.loc[3243:3442,['State', 'County', 'FIPS', 'Adult smoking', 'Excessive drinking']].copy()
````

````python
Alcohol_tobacco.to_csv("Alcohol_tobacco.csv", index=False)
````

### Analysis Area Subset 3

The **third** subset can be used to provide more detail on the third analysis area regarding income and increased funding to a specific area. Although increased funding can be very impactful from a lot of aspects, if we include every column impacted by funding, this subset would be extremely close to the original raw data set. Therefore, we had to limit it down to include only a few critical points. We aimed to cover categories from four different aspects. These categories are:
- children in poverty
- inadequate food
- inadequate water and
- inadequate housing.

There are many more factors affected by funding yet in order to keep this subset brief, we had to limit it to these main aspects.

````python
raw_data.loc[3243:3442,['State', 'County', 'FIPS', 'Children in poverty',  'Drinking water violations', 'Severe housing problems', 'Food insecurity']]
````

````python
Funding = raw_data.loc[3243:3442,['State', 'County', 'FIPS', 'Children in poverty',  'Drinking water violations', 'Severe housing problems', 'Food insecurity']].copy()
````

````python
Funding.to_csv("Funding.csv", index=False)
````
